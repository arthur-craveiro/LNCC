## Notação Assintótica - $\Theta$

```python
from sympy import *
from sympy.plotting import plot
n = symbols('n')
```

```python
g = n*n
g2 = n*n/2
f = n*n/2-2*n
g1 = n*n/4

p = plot(g, f, g1, g2, (n, 0, 50), show=False, legend=True)

#change the color of p2
p[0].line_color = 'k'
p[1].line_color = 'r'
p[2].line_color = 'b'
p.show()
```
