## Questão A
Dado que conhecemos a distribuição de probabilidades de cada variável aleatória e os parâmetros que as caracterizam (Tabela 1), calcule a expectância e a variância (teóricas) de cada uma delas, usando as definições que vimos em aula.


```python
# Q ~ N(3,3) -> A E[Q] = 3 e Var[Q] = 3
print("E[Q] = 3")
print("Var[Q] = 3");

# X ~ U[0,6]:
Xmean = (0+6)/2
Xvar = (6-0)**2/12
print("\nE[X] = ", Xmean)
print("Var[X] =", Xvar)

# Y ~ E(1/3):
Ymean = 1/(1/3)
Yvar = 1/((1/3)**2)
print("\nE[Y] = ", Ymean)
print("Var[Y] =", Yvar)

# T ~ B(10, 0.3):
Tmean = 10*0.3
Tvar = Tmean*(1-0.3)
print("\nE[T] = ", Tmean)
print("Var[T] =", Tvar)
```

### Calculando as médias
* Obs: Usar o sts.mean ou mean. do próprio python resultam no mesmo valor.
* Obs2: Passar a media na variância acelera o seu cálculo (evita recalculos).

```python
Qmean1 = sts.mean(dataQ['dataQ'])
print("E[Q] = ", Qmean1)

Qmean2 = dataQ['dataQ'].mean()
print("E[Q] = ", Qmean2)

Qvar = sts.variance(dataQ['dataQ'], Qmean1)
print("Var[Q] = ", Qvar)
```

```python
Xmean = dataX['dataX'].mean()
print("E[X] = ", Xmean)

Xvar = sts.variance(dataX['dataX'], Xmean)
print("Var[X] = ", Xvar)
```

```python
Ymean = dataY['dataY'].mean()
print("E[Y] = ", Ymean)

Yvar = sts.variance(dataY['dataY'], Ymean)
print("Var[Y] = ", Yvar)
```

```python
Tmean = dataT['dataT'].mean()
print("E[T] = ", Tmean)

Tvar = sts.variance(dataT['dataT'], Tmean)
print("Var[T] = ", Tvar)
```

### Conclusões

1. É possível perceber que no caso de Q (distribuição Normal) e de X (distribuição Uniforme) os erros apresentados são bem pequenos, na casa de 10^(-3) ou menores;

2. Em Y (distribuição Exponencial) o erro da variância aumentou um pouco, na casa de 10^(-2);

3. E em T (distribuição Binomial) o erro também ficou como Q e X, na casa de 10^(-3).

### Para Q ~ N(3, 3):


```python
WQ5, SQ5, Aux = [], [], []
j, i, k, Wi, Wn, ST = 0, 0, 0, 0, 0, 0
n = 5

for j in range(10000):
  for i in range(n):
    Wi = np.random.choice(dataQ['dataQ']) #Sortear de um valor aleatório (VA)
    Aux.append(Wi) #Adicionar ao vetor "auxiliar" que contera todos os Wi momentaneamente (5)

  Wn = sum(Aux)/n #Determinar o valor de um Wn
  WQ5.append(Wn) #Adicionar ao conjunto de VAs Média Amostral

  for k in range (n):
    ST = ST + (Aux[k] - Wn)**2 #Guardar o somatório dos (Wn-Wi) ao quadrado
  Sw = ST/(n-1) #Determinar o valor de um Sw
  SQ5.append(Sw) #Adicionar ao conjunto de VAs Variância Amostral
  Aux = []
  Wn, ST = 0, 0

print(WQ5)
print(SQ5)
```
